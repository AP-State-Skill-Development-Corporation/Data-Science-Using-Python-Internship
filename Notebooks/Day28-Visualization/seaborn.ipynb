 ## creation of subplots
    - plt.subplot()

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1)# row,cols,index/position
plt.plot([1,2,3],[2,3,4])
plt.subplot(1,2,2)
plt.scatter([1,2,3,4,5],[5,6,7,8,9])
```

```python
import numpy as np
plt.subplot(2,1,1)
plt.bar(['x','y','z'],height=5)
plt.subplot(2,1,2)
plt.hist(np.random.randint(1,25,10),bins=5)
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,7,10,15])
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.plot(x,x,'r',label='line plot')
plt.legend()
plt.subplot(2,2,2)
plt.scatter(x,y,label='scatter plot')
plt.legend()
plt.subplot(2,2,3)
plt.bar(x,y,label='bar plot')
plt.legend()
plt.subplot(2,2,4)
plt.barh(x,y,label='scatter plot')
plt.legend()
```

```python
plt.subplot(1,2,1)
plt.pie([20,30,40],labels=['python','java','R'])
plt.subplot(1,2,2)
plt.barh([20,34,49],width=5,color=['r','m','k'],label="bar graph")
plt.legend()
plt.text(4,20,"m1")
plt.text(5,34,"m2")
plt.text(5,49,"m3")
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
plt.plot(np.random.randn(50).cumsum(),'k',label='line plot')
plt.legend()
ax2=fig.add_subplot(2,2,2)
ax2.hist(np.random.randn(100),bins=20,color='y',orientation='horizontal',label="hist plot")
ax3=fig.add_subplot(2,2,3)
ax3.scatter(np.arange(30),np.random.randint(1,50,30),label='scatter plot')
plt.show()
```
