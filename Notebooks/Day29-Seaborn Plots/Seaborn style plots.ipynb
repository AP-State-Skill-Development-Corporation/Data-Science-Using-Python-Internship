## catplot 
- sns.catplot()
    -  interface for drawing categorical plots


```python
sns.catplot(data=df)
```

```python
df.columns
```

```python
sns.catplot(x='sepal_length',y='petal_length',data=df)
```

```python
sns.catplot(x='sepal_length',y='petal_length',data=df,hue='petal_width')
```

```python
sns.catplot(x='sepal_length',y='petal_length',data=df,hue='species')
```

```python
sns.catplot(x='sepal_length',y='species',data=df,hue='species',kind='box')
```

```python
sns.catplot(x='sepal_length',y='species',data=df,hue='species',kind='bar',palette=['r','g','b'])
```

```python
sns.catplot(x='sepal_length',y='species',data=df,kind='bar',palette=['r','g','b'],col='species')
```

```python
sns.catplot(x='sepal_length',y='species',data=df,kind='violin',palette=['r','g','b'],col='species')
```

```python
sns.catplot(x='sepal_length',y='species',data=df,palette=['r','g','b'],marker='^')
```

```python
sns.catplot(x='sepal_length',y='species',kind='boxen',data=df,palette=['r','g','b'])
```

```python
sns.catplot(x='sepal_length',y='species',kind='swarm',data=df,palette=['r','g','b'])
```

## histogram plots
- Plot univariate or bivariate histograms to show distributions of datasets.

- A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

```python
df1=sns.load_dataset("tips")
df1.head()
```

```python
df1.info()
```

```python
df1.columns
```

```python
sns.histplot(data=df1)
```

```python
sns.histplot(x='sex',data=df1)
```

```python
sns.histplot(x='total_bill',data=df1)
```

```python
sns.histplot(x='total_bill',data=df1,bins=5)
```

```python
sns.histplot(y='total_bill',data=df1,binwidth=3)
```

```python
sns.histplot(x='total_bill',data=df1,kde=True)
```

```python
sns.histplot(x='total_bill',data=df1,hue="sex")
```
