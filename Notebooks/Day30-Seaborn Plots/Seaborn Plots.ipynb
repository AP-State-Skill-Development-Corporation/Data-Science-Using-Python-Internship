##  BoX Plot
- seaborn.boxplot()
- Draw a box plot to show distributions with respect to categories.
> A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a
 way that facilitates comparisons between variables or across levels of a categorical variable. 
The box shows the quartiles of the dataset while the whiskers extend to show the rest of 
the distribution, except for points that are determined to be “outliers” using a method that is a 
function of the inter-quartile range.

## kde plots
- kernal density estimator plot
- sns.kdeplot()
- data distributed smothely based on density
- Plot univariate or bivariate distributions using kernel density estimation.

> A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
import seaborn as sns
```

```python
sns.set_theme("poster")
sns.kdeplot(data=df1)
```

```python
sns.kdeplot(x='tip',data=df1)
```

```python
sns.kdeplot(x='tip',y="total_bill",data=df1,hue='sex')
```

```python
sns.kdeplot(x='tip',data=df1,bw_adjust=0.5) #less smoothing
```

```python
sns.kdeplot(x='tip',data=df1,bw_adjust=5,hue='day')
```

```python
sns.kdeplot(x='tip',data=df1,bw_adjust=5,hue='day',multiple='stack')
```

## bar plot()
    - sns.barplot()
> Show point estimates and confidence intervals as rectangular bars.

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. Bar plots include 0 in the quantitative axis range, and they are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.

```python
sns.barplot(data=df1)
```

```python
sns.barplot(x='day',y='tip',data=df1,hue='smoker')
```

```python
import numpy as np
sns.barplot(x='day',y='tip',data=df1,estimator=np.mean)
```

```python
import numpy as np
sns.barplot(x='day',y='tip',data=df1,estimator=np.std)
```

```python
df1['tip'].std()
```

##  point plot
- Show point estimates and confidence intervals using scatter plot glyphs.
- To pointout data or values
- sns.pointplot()

```python
sns.pointplot(data=df1)
```

```python
sns.pointplot(x='sex',y='total_bill',data=df1)
```

```python
sns.pointplot(x='sex',y='total_bill',data=df1,hue='day')
```

## Jointplot
- sns.jointplot()
- Draw a plot of two variables with bivariate and univariate graphs.

```python
sns.set_theme(style='darkgrid')
sns.jointplot(df1['day'],df1['sex'])
```

```python
sns.set_theme(style='darkgrid')
sns.jointplot(df1['total_bill'],df1['tip'],kind='kde')
```

```python
sns.set_theme(style='darkgrid')
sns.jointplot(df1['total_bill'],df1['tip'],kind='reg')
```

```python
sns.set_theme(style='darkgrid')
sns.jointplot(df1['total_bill'],df1['tip'],kind='hex')
```
